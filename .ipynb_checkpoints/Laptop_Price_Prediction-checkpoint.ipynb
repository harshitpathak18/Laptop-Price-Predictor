### PPI - Pixel per inch

```python
# identifying x and y resolution
resolution = df['ScreenResolution'].apply(lambda x:str(x).split()[-1])
x_res = resolution.apply(lambda x:str(x).split('x')[0])
y_res = resolution.apply(lambda x:str(x).split('x')[1])

# typecasting object into float
x_res = x_res.astype('float32')
y_res = y_res.astype('float32')

# calculation ppi
df['ppi'] = (x_res**2 + y_res**2)**0.5/df['Inches']
```

### Touch Screen

```python
df['TouchScreen'] = df['ScreenResolution'].apply(lambda x: 1 if "Touchscreen" in str(x) else 0)
```

### IPS

```python
df['IPS'] = df['ScreenResolution'].apply(lambda x: 1 if "IPS" in str(x) else 0)
```

```python
df.drop(columns=['Inches', 'ScreenResolution'], inplace=True)
```

### CPU Brand

```python
df['CPU Brand'] = df['Cpu'].apply(lambda x: " ".join(str(x).split()[0:3]))
```

```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'

df['CPU Brand'] = df['CPU Brand'].apply(fetch_processor)
```

```python
df.drop(columns=['Cpu'],inplace=True)
```

### Memory

```python
df['Memory'] = df['Memory'].apply(lambda x:str(x).replace('TB','000'))
df['Memory'] = df['Memory'].apply(lambda x:str(x).replace('GB',''))
```

```python
# Adding '+' sign in all those data where + sign is not available so that it can be easily preprocessed
def adding_plus_sign(text):
    if '+' not in text:
        return text + " + "
    else:
        return text
        
df['Memory'] = df['Memory'].apply(adding_plus_sign)
```

```python
# splitting data on '+' sign
first_part = df['Memory'].apply(lambda x: str(x).split("+")[0])
second_part = df['Memory'].apply(lambda x: str(x).split("+")[1])
```

```python
SSD, HDD, Flash_Storage, Hybird = [], [], [], []

for i in range(len(first_part)):
    # Initialize values to 0
    ssd, hdd, flash_storage, hybrid = 0, 0, 0, 0

    value = first_part.iloc[i]
    
    # Check if the value contains 'SSD'
    if 'SSD' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            ssd += int(numeric_part)
    
    # Check if the value contains 'HDD'
    if 'HDD' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            hdd += int(numeric_part)
    
    # Check if the value contains 'Flash Storage'
    if 'Flash Storage' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            flash_storage += int(numeric_part)
    
    # Check if the value contains 'Hybrid'
    if 'Hybrid' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            hybrid += int(numeric_part)

    value = second_part.iloc[i]
    
    # Check if the value contains 'SSD'
    if 'SSD' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            ssd += int(numeric_part)
    
    # Check if the value contains 'HDD'
    if 'HDD' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            hdd += int(numeric_part)
    
    # Check if the value contains 'Flash Storage'
    if 'Flash Storage' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            flash_storage += int(numeric_part)
    
    # Check if the value contains 'Hybrid'
    if 'Hybrid' in value:
        numeric_part = ''.join(filter(str.isdigit, value))  # Extract numeric part
        if numeric_part:  # Check if numeric_part is not empty
            hybrid += int(numeric_part)


    SSD.append(ssd)
    HDD.append(hdd)
    Flash_Storage.append(flash_storage)
    Hybird.append(hybrid)

df['SSD'] = SSD
df['HDD'] = HDD
df['Flash_Storage'] = Flash_Storage
df['Hybrid'] = Hybird
```

### Model Building

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# Splitting features and target
X = df.drop('Price', axis=1)
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)

# Define categorical and numerical features
categorical_features = ['Company', 'TypeName', 'GPU Brand', 'CPU Brand']
numerical_features = ['Ram', 'Weight', 'ppi', 'TouchScreen', 'IPS', 'SSD', 'HDD']

# Create preprocessing steps
categorical_transformer = OneHotEncoder(sparse=False, drop='first')
numerical_transformer = StandardScaler()
```

```python
# Create column transformer
preprocessor = ColumnTransformer(
    transformers=[
        ("Categorical Features", categorical_transformer, categorical_features),
        ("Numerical Features", numerical_transformer, numerical_features)
    ], remainder='passthrough'
)

Models = {"Linear Regression": LinearRegression(), 
          "Ridge": Ridge(), 
          "Lasso": Lasso(),
          "Decision Tree": DecisionTreeRegressor(), 
          "Random Forest": RandomForestRegressor(n_estimators=100,random_state=3, max_samples=0.5,max_features=0.75, max_depth=15), 
          "Gradient Boosting": GradientBoostingRegressor(n_estimators=500), 
          "AdaBoostRegressor": AdaBoostRegressor(n_estimators=15,learning_rate=1.0), 
          "XGBRegressor": XGBRegressor(n_estimators=45,max_depth=8,learning_rate=1.0)}

for name, model in Models.items():
    pipeline = Pipeline(
        steps=[
            ("Preprocessing", preprocessor),
            (f"{name}", model)
        ]
    )
    
    # Fit the pipeline on the training data
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X_test)
    
    # Evaluate the model
    train_score = pipeline.score(X_train, y_train)
    test_score = pipeline.score(X_test, y_test)

    print(f"{name}")
    print(f"Train R2-Score: {train_score}")
    print(f"Test R2-Score: {test_score}")

    print(f"Mean Absolute Error: {mean_absolute_error(y_test, y_pred)}\n\n")
    # # Get feature names after one-hot encoding
    # categorical_feature_names = preprocessor.named_transformers_['Categorical Features'] \
    #                                 .get_feature_names_out(input_features=categorical_features)
    # feature_names = list(categorical_feature_names) + numerical_features
    
    # # Get feature importances (if applicable)
    # if hasattr(model, 'feature_importances_'):
    #     importances = model.feature_importances_
    # elif hasattr(model, 'coef_'):
    #     importances = model.coef_
    # else:
    #     importances = None
    
    # # Print feature names and importances (if available)
    # print(f"{name} Feature Importances:")
    # if importances is not None:
    #     for feature_name, importance in zip(feature_names, importances):
    #         print(f"{feature_name}: {importance}")
    # else:
    #     print("Feature importances not available for this model.")
    # print("\n")
```

### Making Prediction

```python
import pandas as pd
import pickle
```

```python
# Load the model and data
pipe = pickle.load(open('laptop_price_pipeline.pkl','rb'))
df = pd.read_csv('cleaned_laptop_price.csv')

df.head()
```

```python
df.columns
```

```python
def predict_price(company, type, ram, weight, touchscreen, ips, screen_size, resolution, cpu, hdd, ssd, gpu):
    # Convert touchscreen and IPS to binary
    touchscreen = 1 if touchscreen == 'Yes' else 0
    ips = 1 if ips == 'Yes' else 0

    # Calculate PPI
    X_res, Y_res = map(int, resolution.split('x'))
    ppi = ((X_res**2) + (Y_res**2))**0.5 / screen_size

    # Create query array
    # Create DataFrame
    data = pd.DataFrame([[company, type, ram, weight, gpu, ppi, touchscreen, ips, cpu, ssd, hdd]],
                        columns=['Company', 'TypeName', 'Ram', 'Weight', 'GPU Brand', 'ppi', 'TouchScreen', 'IPS', 'CPU Brand', 'SSD','HDD'])

    # Predict price
    predicted_price = pipe.predict(data)[0]
    return int(predicted_price)
```

```python
# Example usage:
predicted_price = predict_price('Dell', 'Gaming', 16, 1.8, 'Yes', 'No', 15.6, '1920x1080', 'Intel Core i5', 512, 0, 'Nvidia')
print("The predicted price of this configuration is", predicted_price)
```
